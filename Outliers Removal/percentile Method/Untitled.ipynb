```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('weight-height.csv')
```

```python
df.sample(5)
```

```python
df.shape
```

```python
df['Height'].describe()
```

```python
plt.figure(figsize = (15,6))
plt.subplot(1,2,1)
sns.distplot(df['Height'])

plt.subplot(1,2,2)
sns.boxplot(df['Height'])

plt.show()
```

```python
upperlimit = df['Height'].quantile(0.99)
upperlimit
```

```python
lowerlimit = df['Height'].quantile(0.01)
lowerlimit
```

```python
new_df = df[(df['Height'] <= upperlimit) & (df['Height'] >= lowerlimit) ]
```

```python
new_df.head()
```

```python
new_df.shape
```

```python
print(f"{new_df['Height'].describe()} \n\n {df['Height'].describe()}")
```

```python
plt.figure(figsize = (15,6))
plt.subplot(1,2,1)
sns.distplot(new_df['Height'])

plt.subplot(1,2,2)
sns.boxplot(new_df['Height'])

plt.show()
```

```python
# Capping --> Winsorization
df['Height'] = np.where(
    df['Height'] >= upperlimit,
    upperlimit,
    np.where(df['Height'] <= lowerlimit,
            lowerlimit,
            df['Height']
            )
)
```

```python
df.shape
```

```python
df['Height'].describe()
```

```python
plt.figure(figsize = (15,6))
plt.subplot(1,2,1)
sns.distplot(df['Height'])

plt.subplot(1,2,2)
sns.boxplot(df['Height'])

plt.show()
```
