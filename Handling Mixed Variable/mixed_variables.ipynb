```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('titanic.csv')
```

```python
df.head()
```

```python
df['number'].unique()
```

```python
fig = df['number'].value_counts().plot.bar()
fig.set_title('Passengers tranvelling with')
```

```python
# Extracting numerical 
df['number_numeric'] = pd.to_numeric(df['number'], errors='coerce', downcast='integer')
```

```python
# Extracting the categorical values
df['number_categoric'] = np.where(df['number_numeric'].isnull(), df['number'], np.nan)
```

```python
df.head()
```

```python
df['Cabin'].unique()
```

```python
df['Ticket'].unique()
```

```python
df['numeric_cabin'] = df['Cabin'].str.extract('(\d+)')
```

```python
df['categoric_cabin'] = df['Cabin'].str[0] # Captures the first letter
```

```python
df.head()
```

```python
df['categoric_cabin'].value_counts().plot(kind='bar')
```

```python
# extract the last bit of ticket as number
df['ticket_num'] = df['Ticket'].apply(lambda s: s.split()[-1])
df['ticket_num'] = pd.to_numeric(df['ticket_num'],
                                   errors='coerce',
                                   downcast='integer')

# extract the first part of ticket as category
df['ticket_cat'] = df['Ticket'].apply(lambda s: s.split()[0])
df['ticket_cat'] = np.where(df['ticket_cat'].str.isdigit(), np.nan,
                              df['ticket_cat'])

df.head(20)
```
