# OneHotEncoding using scikitlearn

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(df.iloc[:, 0:4], df.iloc[:, -1], test_size = 0.2, random_state = False)
```

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
ohe = OneHotEncoder(drop="first",sparse_output = False, dtype = np.int32)
```

```python
X_train_new = ohe.fit_transform(X_train[['fuel', 'owner']])
```

# OneHot Encoding with Top categories

```python
count = df['brand'].value_counts()
```

```python
df['brand'].nunique()
threshold = 100
```

```python
repl = count[count <= threshold].index
```

```python
pd.get_dummies(df['brand'].replace(repl,'uncommon')).sample(5)
```
